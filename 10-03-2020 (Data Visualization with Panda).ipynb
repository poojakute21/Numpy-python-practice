```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.plot(figsize = (10,5))
```

```python
iris.plot(legend= False, figsize = (10,5), logy = True)
```

```python
x = iris.drop(labels = ['sepal_width', 'petal_width'], axis = 1)
x.head()
```

```python
y = iris[['sepal_width', 'petal_width']]
y.head()
```

```python
ax = x.plot()
y.plot(figsize = (16,10), secondary_y = True, ax = ax)
```

# Hex and Pie Plot

```python
df.plot.hexbin(x='sepal_length', y='petal_length', gridsize= 10, C ='sepal_width')
```

```python
df.iloc[0].plot.pie()
```

```python
d = df.head(3).T
d
```

```python
d.plot.pie(subplots= True, figsize =(30,30), fontsize = 24)
plt.show()
```

```python
d.plot.pie(subplots= True, figsize =(30,30), fontsize = 24, autopct = '%.2f')
plt.show()
```

# Scatter Matrix and Subplots

```python
df.plot(subplots = True)
plt.show()
```

```python
df.plot(subplots = True, sharex = False)
plt.tight_layout()
plt.show()
```
